Este conjunto de datos contiene información sobre todos las 802 entradas de la Pokédex de
las primeras siete generaciones de Pokémon. La información contenida en este conjunto de
datos incluye estadísticas base, rendimiento frente a otros tipos, altura, peso, clasificación,
pasos de huevo,puntos de experiencia, habilidades, etc. La informacion se extrajo de
http://serebii.net/Serebi.net

• name: El nombre en inglés del Pokémon.

• japanese name: El nombre japonés original del Pokémon.

• pokedex number: El número de entrada del Pokémon en la Pokédex Nacional.

• percentage male: El porcentaje de especies que son machos. En blanco si el Pokémon
no tiene género.

• type1: El tipo principal de Pokémon.

• type2: El tipo secundario de Pokémon.

• classification: La clasificación de los Pokémon tal como la describe el Pokedex de Sol y
Luna.

• height m: Altura del Pokémon en metros.

• weight kg: El peso del Pokémon en kilogramos.

• capture rate: Tasa de captura del Pokémon.

• baseeggsteps: El número de pasos necesarios para incubar un huevo del Pokémon.

• abilities: Una lista secuenciada de habilidades que el Pokémon es capaz de tener.

• experience growth: El crecimiento de la experiencia del Pokémon.

• base happiness: Felicidad básica del Pokémon.

• against_?: Dieciocho características que denotan la cantidad de daño recibido contra un
ataque de un tipo particular.

• hp: El HP base del Pokémon.

• attack: El ataque base de los Pokémon.

• defense: La defensa básica de los Pokémon.

• sp attack: El ataque especial básico del Pokémon.

• sp defense: La Defensa Especial Base del Pokémon.

• speed: La velocidad base del Pokémon.

• generation: La generación numerada en la que se introdujo por primera vez el Pokémon.

• is legendary: Indica si el Pokémon es legendario.
